```python
from manim import *
import numpy as np

class GeometryProblem5Final(Scene):
    def construct(self):
        r1, r2 = 1.6, 1.1
        c1 = LEFT * 2.5
        c2 = RIGHT * 2.5
        w1 = Circle(radius=r1, color=BLUE).move_to(c1)
        w2 = Circle(radius=r2, color=GREEN).move_to(c2)

        dist = np.linalg.norm(c1 - c2)
        alpha = np.arccos((r1 - r2) / dist)

        A = c1 + r1 * np.array([np.cos(alpha), np.sin(alpha), 0])
        B = c2 + r2 * np.array([np.cos(alpha), np.sin(alpha), 0])
        C = c1 + r1 * np.array([np.cos(-alpha), np.sin(-alpha), 0])
        D = c2 + r2 * np.array([np.cos(-alpha), np.sin(-alpha), 0])

        line_ab = Line(A, B, color=WHITE)
        line_cd = Line(C, D, color=WHITE)

        v_cb = (B - C) / np.linalg.norm(B - C)
        P = C + 2 * np.dot(c1 - C, v_cb) * v_cb

        v_ad = (A - D) / np.linalg.norm(A - D)
        Q = D + 2 * np.dot(c2 - D, v_ad) * v_ad

        ang_p = np.arctan2(P[1]-c1[1], P[0]-c1[0])
        ang_q = np.arctan2(Q[1]-c2[1], Q[0]-c2[0])

        tan_p = TangentLine(w1, alpha=(ang_p % (2*PI))/(2*PI), length=8, color=YELLOW)
        tan_q = TangentLine(w2, alpha=(ang_q % (2*PI))/(2*PI), length=8, color=GOLD)

        S = line_intersection(tan_p.get_start_and_end(), tan_q.get_start_and_end())
        dot_s = Dot(S, color=RED)

        self.play(Create(w1), Create(w2))
        self.play(Create(line_ab), Create(line_cd))

        self.play(Create(Line(C, B, stroke_opacity=0.3)), Create(Line(A, D, stroke_opacity=0.3)))

        self.play(FadeIn(Dot(P, color=YELLOW)), FadeIn(Dot(Q, color=GOLD)))
        self.add(Tex("P", font_size=20).next_to(P, LEFT), Tex("Q", font_size=20).next_to(Q, RIGHT))

        self.play(Create(tan_p), Create(tan_q))
        self.play(FadeIn(dot_s), Write(Tex("S", color=RED).next_to(S, UP)))

        self.play(line_ab.animate.set_color(RED), Indicate(dot_s))
        self.wait(2)
```

```python
from manim import *
import numpy as np

class GeometryProblem5FullSolution(Scene):
    def construct(self):
        r1, r2 = 1.6, 1.1
        c1, c2 = LEFT * 2.5, RIGHT * 2.5
        w1 = Circle(radius=r1, color=BLUE).move_to(c1)
        w2 = Circle(radius=r2, color=GREEN).move_to(c2)

        dist = np.linalg.norm(c1 - c2)
        alpha = np.arccos((r1 - r2) / dist)
        A = c1 + r1 * np.array([np.cos(alpha), np.sin(alpha), 0])
        B = c2 + r2 * np.array([np.cos(alpha), np.sin(alpha), 0])
        C = c1 + r1 * np.array([np.cos(-alpha), np.sin(-alpha), 0])
        D = c2 + r2 * np.array([np.cos(-alpha), np.sin(-alpha), 0])

        v_cb = (B - C) / np.linalg.norm(B - C)
        P = C + 2 * np.dot(c1 - C, v_cb) * v_cb
        v_ad = (A - D) / np.linalg.norm(A - D)
        Q = D + 2 * np.dot(c2 - D, v_ad) * v_ad

        ang_p = np.arctan2(P[1]-c1[1], P[0]-c1[0])
        ang_q = np.arctan2(Q[1]-c2[1], Q[0]-c2[0])
        tan_p = TangentLine(w1, alpha=(ang_p % (2*PI))/(2*PI), length=10, color=YELLOW)
        tan_q = TangentLine(w2, alpha=(ang_q % (2*PI))/(2*PI), length=10, color=GOLD)
        S = line_intersection(tan_p.get_start_and_end(), tan_q.get_start_and_end())

        formula1 = MathTex(r"AB^2 = BP \cdot BC = AQ \cdot AD", font_size=30).to_edge(UP)
        formula2 = MathTex(
            r"\frac{AQ}{\sin \angle AM'Q} = \frac{BP}{\sin \angle PMB}",
            font_size=30
            ).next_to(formula1, DOWN)

        conclusion = MathTex(r"AM' = PM \Rightarrow M = M'", color=RED, font_size=36).to_edge(DOWN)

        self.play(Create(w1), Create(w2))
        self.play(Create(Line(A, B)), Create(Line(C, D)))

        self.play(Write(formula1))
        self.play(Create(Line(C, B, stroke_opacity=0.4)), Create(Line(A, D, stroke_opacity=0.4)))
        self.play(FadeIn(Dot(P)), FadeIn(Dot(Q)))

        self.play(Write(formula2))
        self.play(Create(tan_p), Create(tan_q))

        self.play(FadeIn(Dot(S, color=RED)), Write(Tex("S/M", color=RED, font_size=24).next_to(S, UP)))
        self.play(Write(conclusion))
        self.play(Indicate(Line(A, B), color=RED))

        self.wait(3)
```

```python
from manim import *
import numpy as np

class Solution(Scene):
    def construct(self):
        r1, r2 = 1.6, 1.1
        c1_pos, c2_pos = LEFT * 2.5, RIGHT * 2.5
        w1 = Circle(radius=r1, color=BLUE).move_to(c1_pos)
        w2 = Circle(radius=r2, color=GREEN).move_to(c2_pos)

        dist = np.linalg.norm(c1_pos - c2_pos)
        alpha = np.arccos((r1 - r2) / dist)
        A = c1_pos + r1 * np.array([np.cos(alpha), np.sin(alpha), 0])
        B = c2_pos + r2 * np.array([np.cos(alpha), np.sin(alpha), 0])
        C = c1_pos + r1 * np.array([np.cos(-alpha), np.sin(-alpha), 0])
        D = c2_pos + r2 * np.array([np.cos(-alpha), np.sin(-alpha), 0])

        v_cb = (B - C) / np.linalg.norm(B - C)
        P = C + 2 * np.dot(c1_pos - C, v_cb) * v_cb
        v_ad = (A - D) / np.linalg.norm(A - D)
        Q = D + 2 * np.dot(c2_pos - D, v_ad) * v_ad

        ang_p = np.arctan2(P[1]-c1_pos[1], P[0]-c1_pos[0])
        ang_q = np.arctan2(Q[1]-c2_pos[1], Q[0]-c2_pos[0])
        tan_p = TangentLine(w1, alpha=(ang_p % (2*PI))/(2*PI), length=10, color=YELLOW)
        tan_q = TangentLine(w2, alpha=(ang_q % (2*PI))/(2*PI), length=10, color=GOLD)
        M = line_intersection(tan_p.get_start_and_end(), tan_q.get_start_and_end())

        title = Text(" 5:  ", font_size=24).to_edge(UP)
        f1 = MathTex(r"AB^2 = BP \cdot BC = AQ \cdot AD \Rightarrow AQ = BP", font_size=26).to_edge(UP, buff=0.8)
        f2 = MathTex(
            r"\frac{AQ}{\sin \angle AM'Q} = \frac{BP}{\sin \angle PMB}",
            font_size=26
            ).next_to(f1, DOWN, buff=0.2)

        labels = VGroup(
            Tex("A", font_size=20).next_to(A, UP),
            Tex("B", font_size=20).next_to(B, UP),
            Tex("C", font_size=20).next_to(C, DOWN),
            Tex("D", font_size=20).next_to(D, DOWN),
            Tex("P", font_size=20).next_to(P, LEFT),
            Tex("Q", font_size=20).next_to(Q, RIGHT)
            )

        self.play(Write(title))
        self.play(Create(w1), Create(w2))
        self.play(Create(Line(A, B)), Create(Line(C, D)), FadeIn(labels))

        self.play(Write(f1))
        self.play(Create(Line(C, B, stroke_opacity=0.3)), Create(Line(A, D, stroke_opacity=0.3)))
        self.play(FadeIn(Dot(P)), FadeIn(Dot(Q)))
        self.wait(0.5)

        self.play(Write(f2))
        self.play(Create(tan_p), Create(tan_q))

        dot_m = Dot(M, color=RED)
        self.play(FadeIn(dot_m), Write(Tex("M", color=RED).next_to(M, UP)))
        self.play(Indicate(Line(A, B), color=RED), Flash(dot_m, color=RED))
        self.wait(3)
```
